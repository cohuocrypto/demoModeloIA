#1. Estructuras Nativas de Python
## 1.1 Tuplas (inmutables)

### Las tuplas se utilizan para almacenar datos que no deben cambiar durante la ejecución del programa.

```python
# Definición de una tupla (Coordenadas de un centro de datos)
ubicacion_server = (40.7128, -74.0060)

print(f"Latitud: {ubicacion_server[0]}")
# ubicacion_server[0] = 40.7200  # <--- Esto daría ERROR porque son inmutables
```

## 1.2 Diccionarios {'clave': 'valor'}

### Fundamentales en IA para manejar hiperparámetros de modelos o configuraciones.

```python
# Configuración de un modelo de Machine Learning
hiperparametros = {
    "learning_rate": 0.01,
    "epochs": 100,
    "optimizer": "Adam",
    "layers": [64, 32, 16]
}

print(f"El optimizador seleccionado es: {hiperparametros['optimizer']}")
```

### 1.3 List Comprehensions
Una forma eficiente de transformar datos en una sola línea.

```python
# Supongamos que tenemos errores de predicción y queremos elevarlos al cuadrado (MSE)
errores = [1.2, -0.5, 2.0, -1.1]
errores_cuadrados = [x**2 for x in errores]

print(f"Errores al cuadrado: {errores_cuadrados}")
```

##  2. Estructuras Científicas (NumPy)
Para usar estas estructuras, necesitamos importar la librería numpy. Son la base del Deep Learning.

## import numpy as np

```python
import numpy as np
```

## 2.1 Vectores (Arrays 1D)
Representan una sola característica (feature) en un dataset.

```python
# Vector de pesos iniciales de una neurona
pesos = np.array([0.5, -0.2, 0.1])
print(f"Vector de pesos: {pesos}")
```

## 2.2 Matrices (Arrays 2D)
Representan el conjunto de datos de entrenamiento (X).

```python
# Matriz de 3 ejemplos con 2 características cada uno
# Filas = Ejemplos, Columnas = Características (ej: Edad, Salario)
data_input = np.array([
    [25, 30000],
    [30, 45000],
    [40, 80000]
])

print("Matriz de datos:")
print(data_input)
print(f"Forma de la matriz (filas, columnas): {data_input.shape}")
```

## 3. Estructuras para Análisis de Datos (Pandas)
El DataFrame es la herramienta estándar para manipular datos tabulares.

```python
import pandas as pd

# Crear un DataFrame desde cero (Simulando un dataset de clientes)
data = {
    'Cliente_ID': [1, 2, 3],
    'Edad': [25, 30, 40],
    'Compra_Total': [150.50, 200.00, 50.75],
    'Premium': [True, False, True]
}

df = pd.DataFrame(data)

# Visualización
print("Estructura del DataFrame:")
display(df.head()) # En Notebooks 'display' genera una tabla visual muy limpia

# Operación rápida: Media de compras
print(f"\nCompra promedio: {df['Compra_Total'].mean()}")
```
